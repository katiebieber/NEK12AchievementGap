```r
library(dplyr)
library(ggplot2)
library(lubridate)
```

Read csv file into data frame.

```r
Gradraw <- read.csv("StateLevelData.csv")
```

Filter out Native Hawaiian field as there are less than 10 in the state so that field is masked (throws an error if you don't filter).

```r
Grad <- Gradraw %>% filter(Race!="Native Hawaiian or Other Pacific Islander") 
```

Create another column for Graduation Percent and turn the column into raw numbers not alpha characters like CohortGradRate column is.

```r
Grad$GradPerc <- as.numeric(as.character(Grad$CohortGradRate))
```

Read data head.

```r
head(Grad)
```

Filter FRL column as FRL only.

```r
gradFRLRace <- Grad %>%
  filter(FRL == "FRL") 
```

Plot line graph of free and reduced price lunch by race, females vs males, with the x as the data year and the y as the graduate percentage. Do a facet wrap to include a separate line graph for each gender. Set width and height of each graph to 10x10.

```r
library(repr)
options(repr.plot.width=10, repr.plot.height=10)
ggplot(data=gradFRLRace, aes(x=Datayear, y=GradPerc, group=Gender, color=Gender)) + geom_line() + labs(x="School Year", y="Graduation Rate", title="Graduation Rate by Free or Reduced Price Lunch by Race", subtitle="Females vs. Males", caption="Graphic by Katie Bieber") + facet_wrap(~Race, scales='free_x') + theme(axis.line=element_line()) + scale_x_continuous(labels=as.character(gradFRLRace$Datayear),breaks=gradFRLRace$Datayear)
```

Export file as pdf set with same 10x10 width and height.

```r
ggsave("gradFRLRace.pdf", width=10, height=10)
```
